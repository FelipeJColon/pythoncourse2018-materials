### Figure, Axes, Axis

* **Figure** is the whole image, a 'container' that holds all objects of your final image.
* **Figure** can have any number of **Axes**, but to be useful should have at least one.
* **Axes** is the region of a **Figure** that displays your data.
* **Axes** contains two (or three in the case of 3D) **Axis** objects which control the data limits.

```python
fig = plt.figure() # an empty Figure with no Axes (creates a figure object)
```

```python
fig, axes = plt.subplots() # an empty Figure with one Axes that contains two Axis
```

Lets draw two cosine functions of different amplitude on the same **Axes**.

```python
import numpy as np
```

```python
# Create data
X = np.linspace(-np.pi, np.pi, 100, endpoint=True)
C1 = np.cos(X)
C2 = 2*np.cos(X)
```

```python
# Plot data
fig, axes = plt.subplots()
axes.plot(X, C1);
axes.plot(X, C2);
```

**Exersise 3**. Lets move the center of the coordinate system (the origin) to (0,0). For that we need to work with the **spines**, lines connecting the axis tick marks and noting the boundaries of the data area.
1. Use axes.spines(). 
2. Hide two out of four spines (say, top and right) by using .set_visible().
3. Adjuct the other two spines (say, bottom and left) to intersect at (0,0) by using set_position().

```python
# Move origin to (0,0)
fig, axes = plt.subplots()
axes.plot(X, C1);
axes.plot(X, C2);
axes.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]);
axes.set_yticks([-2, -1, 0, 1, 2]);
axes.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']);

axes.spines['right'].set_visible(False)
axes.spines['top'].set_visible(False)
axes.spines['bottom'].set_position(('data',0))
axes.spines['left'].set_position(('data',0))
```

```python
# Solution
# Add legend
fig, axes = plt.subplots()
axes.plot(X, C1, label='cos(x)');
axes.plot(X, C2, label='2cos(x)');
axes.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]);
axes.set_yticks([-2, -1, 0, 1, 2]);
axes.set_xticklabels([r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']);
axes.spines['right'].set_visible(False)
axes.spines['top'].set_visible(False)
axes.spines['bottom'].set_position(('data',0))
axes.spines['left'].set_position(('data',0))

axes.legend(loc='upper left', frameon=False);
```

# Example - Contour plots

```python
#### Create a function for generating visual data
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
```

```python
### Creates data
n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
```

```python
### Object orientated 
fig4 = plt.figure()
ax = fig4.add_subplot(111)

### Plot a contour plot
ax.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap=plt.cm.hot)
C = ax.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)

### Do you want values inline with the contours? If so, uncomment me below!
#ax.clabel(C, inline=1, fontsize=30)
print("Did you label the values inline?")
```

# Example 2 -  Quiver Plots

```python
### Creates random data 'X' 'Y' 'U' 'V'
n = 8
X,Y = np.mgrid[0:n,0:n]
T = np.arctan2(Y-n/2.0, X-n/2.0)
U,V = R*np.cos(T), R*np.sin(T)
```

```python
### Creates some data that we will use for the colouring of the arrows.
R = 10+np.sqrt((Y-n/2.0)**2+(X-n/2.0)**2)
```

```python
# Plots using object orientated way
fig6 = plt.figure()
ax = fig6.add_subplot(111)

# Make a quiver plot. and colour is using 'R'
ax.quiver(X,Y,U,V,R, alpha=.5)

### If you'd like to make it look a little more perfected and better contrast, try this one: 
#ax.quiver(X,Y,U,V, facecolor='None', linewidth=.5)
 
# Set the limits.
ax.set_xlim(-1,n)
ax.set_ylim(-1,n)
```
